# Solution to CRM ODE using DifferentialEquations.jl

Timestep #300-301

```julia
function crm_bhp!(du, u, p, t)
    q = u
    τ, J, F1, F2, F3, F4, I1, I2, I3, I4, δp = p
    # I1 = (df[1:end, 2])
    # I2 = (df[1:end, 3])
    # I3 = (df[1:end, 4])
    # I4 = (df[1:end, 5])
    du[1] = ((I1.*F1)+(I2.*F2)+
        (I3.*F4)+(I4.*F4) .- (τ.*J.*δp) .- u[1])./τ 
end
```

```julia
function parameterized_crm_bhp!(du, u, p, t)
    I1 = (df[1:end, 2])
    I2 = (df[1:end, 3])
    I3 = (df[1:end, 4])
    I4 = (df[1:end, 5])
    δp = (df[1:end, end-1])
    p_ = (p[1], p[2], p[3], p[4], p[5], p[6], I1[300], I2[300], I3[300], I4[300], δp[300])
    crm_bhp!(du, u, p_, t)
end
```

```julia
u0 = [df[300, :PA01]]
tspan = (300, 301)
par = [21.3, 18.6, 0.372, 0, 0.205, 0.205]
# true_param = (p[1], 
#     I1[300], I2[300], I3[300], I4[300], δp[300])
prob1 = ODEProblem(parameterized_crm_bhp!, u0, tspan, par, saveat=0.1)
sol1 = solve(prob1)
scatter(sol1, xlabel="t", ylabel="q")
```

```julia
I1 = (df[1:end, 2])
I2 = (df[1:end, 3])
I3 = (df[1:end, 4])
I4 = (df[1:end, 5])
δp = (df[1:end, end-1])
prob2 = ODEProblem(crm_bhp!, u0, tspan, 
    (par[1], par[2], par[3], par[4], par[5], par[6], I1[300], I2[300], I3[300], I4[300], δp[300]),
    saveat=0.1)
sol2 = solve(prob2)
scatter(sol2, xlabel="t", ylabel="q")
```

```julia
N = 10
# initial_conditions = df[300, :PA01]
new_tspan = hcat(df[300:309, :Time], df[301:310, :Time])
p_ = hcat(I1[300:309], I2[300:309], I3[300:309], I4[300:309], δp[300:309])
function prob_func(prob, i, repeat)
    # remake(prob,u0=initial_conditions[i])
    p = (prob.p[1], prob.p[2], prob.p[3], prob.p[4], prob.p[5], prob.p[6], p_[i], p_[i+N], p_[i+2*N], p_[i+3*N], p_[i+4*N])
    ODEProblem(
        prob2.f,
        prob2.u0,
        (new_tspan[i], new_tspan[i+N]),
        p)
end
monte_prob = EnsembleProblem(prob2,prob_func=prob_func)
```

```julia
# when using sciml_train, the first return is the loss value,
# and the other returns are sent to the callback for monitoring convergence.
function loss(p)
    sol = solve(prob1, p=p, saveat=0.1)
    loss = sum(abs2, sol.-1)
    return loss, sol
end
```

```julia
callback = function (p, l, pred)
  display(l)
  plt = plot(pred, ylim = (0, 6))
  display(plt)
  # Tell sciml_train to not halt the optimization. If return true, then
  # optimization stops.
  return false
end
```
